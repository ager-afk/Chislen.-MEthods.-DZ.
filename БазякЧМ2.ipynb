```python
# Импорт необходимых библиотек
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
import math

# Настройка отображения графиков
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
```

```python
import numpy as np

def f(x):
    """Исходная функция."""
    return 2**x - 2 * np.cos(x)

def phi(x):
    """
    Итерационная функция φ(x) = log₂(|2cos(x)|)
    Берем модуль, чтобы избежать отрицательных значений под логарифмом.
    """
    arg = 2 * np.cos(x)
    return np.log2(np.abs(arg) + 1e-15)  # добавляем маленькое число для защиты от нуля

def simple_iteration_method(x0, eps, max_iter=100):
    """
    Находит корень уравнения x = phi(x) методом простой итерации.
    """
    print("Таблица итераций метода простой итерации:")
    print("┌─────┬───────────┬───────────┬───────────┐")
    print("│  n  │     x_n   │  phi(x_n) │ |x_n+1 - x_n| │")
    print("├─────┼───────────┼───────────┼───────────┤")

    x_prev = x0
    for n in range(max_iter):
        x_next = phi(x_prev)
        diff = abs(x_next - x_prev)
        
        print(f"│ {n+1:3d} │ {x_prev:9.6f} │ {x_next:9.6f} │ {diff:9.6f} │")

        if diff < eps:
            print("└─────┴───────────┴───────────┴───────────┘")
            print(f"\nРезультат:")
            print(f"Корень x = {x_next:.6f}")
            print(f"Значение функции f(x) = {f(x_next):.6f}")
            print(f"Число итераций: {n+1}")
            return x_next
        
        x_prev = x_next

    print("└─────┴───────────┴───────────┴───────────┘")
    print(f"Достигнуто максимальное число итераций ({max_iter}).")
    return x_prev

# Запуск метода
x0 = 11.0
eps = 0.001
root_iteration = simple_iteration_method(x0, eps, max_iter=100)
```
