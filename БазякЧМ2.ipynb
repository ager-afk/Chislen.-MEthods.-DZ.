Практическая работа №3: Решение уравнений методом итераций

```python
from numpy import *
import pandas as pd

# 1. Исходная функция
def f(x):
    return 2**x - 2 * cos(x)

# 2. Итерационная функция x = x - tau * f(x)
def phi(x, tau):
    return x - tau * f(x)

def simple_iteration_method(x0, eps, max_iter=100):
    # Параметр tau подбирается для сходимости. 
    # Для этого уравнения на интервале [0, 1] оптимально 0.3
    tau = 0.3
    
    history = []
    x_prev = x0
    
    for n in range(max_iter):
        x_next = phi(x_prev, tau)
        diff = abs(x_next - x_prev)
        
        # Сохраняем данные для таблицы
        history.append({
            'n': n + 1,
            'x_n': round(x_prev, 6),
            'phi(x_n)': round(x_next, 6),
            '|x_n+1 - x_n|': round(diff, 6)
        })

        if diff < eps:
            break
        x_prev = x_next

    # Формирование и вывод таблицы через Pandas
    df = pd.DataFrame(history)
    print("Таблица итераций (Метод простой итерации):")
    print(df.to_string(index=False))
    
    print("-" * 50)
    print(f"Результат:")
    print(f"Корень x = {x_prev:.6f}")
    print(f"Значение f(x) = {f(x_prev):.6f}")
    print(f"Число итераций: {len(history)}")
    
    return x_prev

# Параметры запуска
x_start = 0.5   # Начальное приближение на отрезке [0, 1]
precision = 0.001
root = simple_iteration_method(x_start, precision)
```
