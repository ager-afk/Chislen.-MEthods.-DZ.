Решение упражнений с измененными именами переменных

# Упражнение 1 - Решение уравнения lg x - cos x = 0 методом половинного деления

```python
from numpy import *
import pandas as pd

# 1. Определение функции (log10 в numpy это десятичный логарифм)
def f(x):
    return log10(x) - cos(x)

# 2. Параметры
a = 1.0
b = 2.0
eps = 0.001

# Список для хранения итераций
data = []

# 3. Метод половинного деления
def bisection(a, b, eps):
    n = 0
    while (b - a) / 2 > eps:
        n += 1
        c = (a + b) / 2
        f_c = f(c)
        
        data.append({
            'Итер.': n,
            'a': round(a, 4),
            'b': round(b, 4),
            'x (середина)': round(c, 4),
            'f(x)': round(f_c, 4),
            'Погрешность': round((b - a) / 2, 4)
        })
        
        if f(a) * f_c < 0:
            b = c
        else:
            a = c
            
    return (a + b) / 2

# Запуск расчета
root = bisection(a, b, eps)

# 4. Вывод таблицы и результата
df = pd.DataFrame(data)
print("Метод половинного деления для lg(x) - cos(x) = 0:")
print(df.to_string(index=False))
print("-" * 50)
print(f"Приблизительный корень: x ≈ {root:.4f}")
```

# Упражнение 2 - Нахождение корней квадратного уравнения x^2 - 3.2x - 1 = 0

```python
from numpy import *
import pandas as pd

def f(x):
    return x**2 - 3.2*x - 1

def df(x):
    return 2*x - 3.2

# Параметры
x_curr = 4.0
eps = 0.0001
data = []
n = 0

while True:
    n += 1
    x_next = x_curr - f(x_curr) / df(x_curr)
    diff = abs(x_next - x_curr)
    
    data.append({
        'Итерация': n,
        'x_n': round(x_curr, 6),
        'f(x_n)': round(f(x_curr), 6),
        'Погрешность': round(diff, 6)
    })
    
    if diff < eps:
        break
    x_curr = x_next

# Вывод результатов
df_res = pd.DataFrame(data)
print("Метод Ньютона для x^2 - 3.2x - 1 = 0:")
print(df_res.to_string(index=False))
print(f"\nУточненный корень: x ≈ {x_next:.5f}")
```

# Упражнение 3 - Решение уравнения 4x - 3ln x = 4 методом простой итерации

```python
from numpy import *
import pandas as pd

# 1. Функция f(x)
def f(x):
    return 4*x - 3*log(x) - 4

# 2. Функция итераций phi(x) = (3*ln(x) + 4) / 4
def phi(x):
    return (3*log(x) + 4) / 4

# Параметры
x0 = 1.1 # Начальное приближение чуть правее корня
eps = 0.001
data = []
x_curr = x0
n = 0

# 3. Итерационный процесс
while True:
    n += 1
    x_next = phi(x_curr)
    diff = abs(x_next - x_curr)
    
    data.append({
        'Итерация': n,
        'x_n': round(x_curr, 5),
        'x_{n+1}': round(x_next, 5),
        '|x_{n+1}-x_n|': round(diff, 5)
    })
    
    if diff < eps or n > 50:
        break
    x_curr = x_next

# 4. Вывод результатов
df = pd.DataFrame(data)
print(df.to_string(index=False))
print(f"\nНаибольший положительный корень: x ≈ {x_next:.4f}")
```
