# Упражнение 1 - Решение уравнения lg x - cos x = 0 методом половинного деления

```python
from math import *
from matplotlib.pyplot import *

def funkciya_1(x):
    """Функция f(x) = lg x - cos x"""
    return log10(x) - cos(x)

# Создаем массив точек для построения графика
x_koord = [0.1 + i * (4 - 0.1) / 499 for i in range(500)]
y_koord = [funkciya_1(xi) for xi in x_koord]

# Строим график функции
figure(1)
plot(x_koord, y_koord)
grid(True)
axhline(y=0, color='r')
title('lg x - cos x = 0')
xlabel('x')
ylabel('f(x)')
show()

# Графически определяем, что корень находится между 1 и 2
levaya_granica, pravaya_granica = 1, 2
tochnost = 0.00005  # точность 10^-5 для 4 знаков после запятой
nomer_iteracii = 0

print("--- Метод половинного деления для уравнения lg x - cos x = 0 ---")
print(f"Начальный отрезок: [{levaya_granica}, {pravaya_granica}]")
print(f"Точность: {tochnost}")
print("-" * 50)

# Цикл половинного деления
while pravaya_granica - levaya_granica > tochnost:
    nomer_iteracii += 1
    seredina = (levaya_granica + pravaya_granica) / 2
    print(f"Итерация {nomer_iteracii:2d}: x = {seredina:.5f}")
    
    if funkciya_1(levaya_granica) * funkciya_1(seredina) <= 0:
        pravaya_granica = seredina
    else:
        levaya_granica = seredina

koren_1 = (levaya_granica + pravaya_granica) / 2
print("-" * 50)
print(f"Количество итераций: {nomer_iteracii}")
print(f"Найденный корень: {koren_1:.5f}")
print(f"Проверка: f({koren_1:.5f}) = {funkciya_1(koren_1):.8f}")
print()
```

# Упражнение 2 - Нахождение корней квадратного уравнения x^2 - 3.2x - 1 = 0

```python
def funkciya_2(x):
    """Функция f(x) = x^2 - 3.2x - 1"""
    return x**2 - 3.2*x - 1

# Создаем точки для графика без numpy
x_znacheniya = [-2 + i * (5 - (-2)) / 1000 for i in range(1001)]
y_znacheniya = [funkciya_2(x) for x in x_znacheniya]

# Строим график
figure(figsize=(12, 8))
plot(x_znacheniya, y_znacheniya, 'r-', linewidth=2, label='f(x) = x² - 3.2x - 1')
axhline(y=0, color='k', linestyle='-', alpha=0.3)
axvline(x=0, color='k', linestyle='-', alpha=0.3)
grid(True, alpha=0.3)
xlabel('x')
ylabel('f(x)')
title('Графическое отделение корней квадратного уравнения')
legend()

# Находим точные корни через дискриминант
diskriminant = 3.2**2 + 4
koren_2_1 = (3.2 + sqrt(diskriminant)) / 2
koren_2_2 = (3.2 - sqrt(diskriminant)) / 2
print("--- Квадратное уравнение x² - 3.2x - 1 = 0 ---")
print(f"Точные корни: {koren_2_1:.4f} и {koren_2_2:.4f}")

# Отмечаем корни на графике
plot(koren_2_1, 0, 'bo', markersize=8, label=f'Корень 1: x ≈ {koren_2_1:.3f}')
plot(koren_2_2, 0, 'go', markersize=8, label=f'Корень 2: x ≈ {koren_2_2:.3f}')
legend()
show()
print()
```

# Упражнение 3 - Решение уравнения 4x - 3ln x = 4 методом простой итерации

```python
def funkciya_3(x):
    """Исходная функция f(x) = 4x - 3*ln(x) - 4"""
    return 4*x - 3*log(x) - 4

def iteracionnaya_funkciya(x):
    """Итерационная функция x = (4 + 3*ln(x))/4"""
    return (4 + 3*log(x)) / 4

# Создаем точки для графика
x_koord = [0.1 + i * (4 - 0.1) / 499 for i in range(500)]
y_koord = [funkciya_3(x) for x in x_koord]

# Находим корень методом простой итерации
x_tekushiy = 2.0  # начальное приближение
tochnost_iter = 1e-4
max_iteraciy = 100

print("--- Метод простой итерации для уравнения 4x - 3ln x = 4 ---")
print(f"Начальное приближение: x₀ = {x_tekushiy}")
print(f"Точность: {tochnost_iter}")
print("-" * 50)

for nomer in range(max_iteraciy):
    x_noviy = iteracionnaya_funkciya(x_tekushiy)
    if abs(x_noviy - x_tekushiy) < tochnost_iter:
        koren_3 = x_noviy
        print(f"Итерация {nomer+1}: x = {x_noviy:.6f} (достигнута точность)")
        break
    print(f"Итерация {nomer+1}: x = {x_noviy:.6f}")
    x_tekushiy = x_noviy

print("-" * 50)
print(f"Найденный корень: {koren_3:.6f}")
print(f"Проверка: f({koren_3:.6f}) = {funkciya_3(koren_3):.8f}")
print()

# Строим график функции
figure(figsize=(12, 6))
plot(x_koord, y_koord, 'b-', linewidth=2, label='f(x) = 4x - 3·ln(x) - 4')
axhline(0, color='k', linestyle='-')
axvline(0, color='gray', alpha=0.5)
grid(True, alpha=0.3)

# Отмечаем найденный корень
plot(koren_3, 0, 'ro', markersize=8, label=f'корень: x = {koren_3:.4f}')
plot(koren_3, funkciya_3(koren_3), 'ro', markersize=8)

# Оформление графика
title('Решение уравнения 4x - 3·ln x = 4')
xlabel('x')
ylabel('f(x)')
legend()
xlim(0, 3)
ylim(-3, 5)
show()
```
